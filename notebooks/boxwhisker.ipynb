```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Generate datasets
np.random.seed(0)  # For reproducibility

# Dataset 1: Small spread
data_small_spread = np.random.normal(loc=80, scale=1, size=100)

# Dataset 2: Large spread
data_large_spread = np.random.normal(loc=80, scale=5, size=100)

# Combine datasets into a dictionary for easy plotting
data = {
    'Homogeneous': data_small_spread,
    'Not Homogeneous': data_large_spread
}

# Create side-by-side boxplots with custom colors
sns.boxplot(data=data, palette=['#CD00EF', '#24E9CB'])

# Add title and labels
plt.title('Distribution of Similarity Scores')
plt.xlabel('Bag')
plt.ylabel('Similarity Score')

# Add red dotted line
plt.axhline(75, color='red', linestyle='dotted')
plt.text(0.5,74, 'Threshold', color='red', ha='center')

# Show the plot
plt.show()
```
